```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
```

```python
x = tf.Variable(-1.0)
y = tf.Variable(2.0)
```

```python
with tf.GradientTape() as t:
    f9=x+y
    f6=f9**3
    f7=f9**2
    f8=3*y
    f5=f6+f7
    f4=2*x
    f3=f4+f5
    f2=f3+f8
    f1=f2+1
    tf.print(f"the value of f(x,y) at (-1,2): {f1:4.3}")
g=t.gradient(f1,[x,y])
tf.print(f"gradient  df/dx: {g[0]:4.3}, df/dy:  {g[1]:4.3}")
```
