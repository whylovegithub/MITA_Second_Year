```python
import tensorflow as tf
print("Tensorflow version:", tf.__version__)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# start assigning value to w.
np.random.seed(555)
w_1 = np.random.normal(loc=1, scale=0.3, size=3)
print(w_1)
w_1[0] = 0.1

```

```python
# start generating value to matrix x.
# setting sample size
w=w_1
n=500
x = [np.random.uniform(low=0, high=1, size=2) for i in range(n)]

#forming y
y=[-1 if w[0]+w[1]*x[0]-w[2]*x[1]<0 else 1 for x in x]

w_2 = w_1

w_2[0] = 0.1
w_2[1] = -1.2
w_2[2] = 0.3
```

```python
w=w_2

maxIter = 15
error = 0.001
alpha = 0.01
#b=tf.constant(b,dtype='float32')
vector = np.vectorize(np.float32)
x = vector(x)
w = vector(w)
y = vector(y)


# loss=0
# for i in range(n):
#     temp = -1*y[i]*x[i]
#     temp2 = temp[0]*w[0]+temp[1]*w[1]-temp[2]*w[2]
#     loss = loss + tf.math.log(tf.exp(temp2)+1)
# print(loss)

f1=0
z=0



while z in range(maxIter):
    print("we are in ",z+1," iteration")
    w1 = tf.Variable(w[1])
    w2 = tf.Variable(w[2])
    #wVector = np.array([np.array(w1),np.array(w2)])
    tf.print("current w1 & w2:",w1,w2)
    with tf.GradientTape(persistent=True) as t:
        with tf.GradientTape() as tt:
            f13 = w1**2
            f14 = w2**2
            f12 = f13+f14
            f11 = alpha*f12
            f10 = tf.multiply(x[:,0],w1)
            f9 = -1*tf.multiply(x[:,1],w2)
            f8 = w[0]
            f6 = f8+f9+f10
            f5 = tf.multiply(-y,f6)
            f4 = tf.exp(f5)
            f3 = 1+f4
            f2 = tf.math.log(f3)
            f1 = tf.reduce_sum(f2)+tf.reduce_sum(f11)
            
            
        g=tt.gradient(f1,[w1,w2])
    H1=t.gradient(g[0],[w1,w2])
    H2=t.gradient(g[1],[w1,w2]) 
    h = np.array([np.array(H1),np.array(H2)])
    h = vector(h)
    tf.print("Hessian H first row:\n", H1)
    tf.print("Hessian H second row:\n", H2)
    gT = np.array(g).transpose()
    tf.print("gradient gT:\n", gT)
    a = np.linalg.solve(h,gT)
    tf.print("eigenalues of h:\n",np.linalg.eig(h))
    tf.print("norm of g:\n",np.linalg.norm(g))
    tf.print("newtown direction :\n", a)
    print("slope:",w[1],"/",w[2],"  ",w[1]/w[2],"  b:",w[0]/w[2])
    fig, ax = plt.subplots()
    c = ['orange' if item == -1 else 'blue' for item in y]  
    ax.set_xlim((-1, 1))
    ax.set_ylim((0, 1))
    ax.scatter(*zip(*x), c = c, s = 1)
    xlinspace = np.linspace(-1, 1, 400)  
    print(w[0],w[1],w[2])
    classcifier = w[1]/w[2]*xlinspace + w[0]/w[2]
    ax.plot(xlinspace, classcifier)
    plt.show()
    w[1]=w[1]-a[0]
    w[2]=w[2]-a[1]
    z=z+1
    print("loss: ",f1,"\n")
    f1=0
    print("\n\n\n")
    if np.linalg.norm(g)<error:
        print("satisfied")
        break
    
    
    
 


    
    
```
